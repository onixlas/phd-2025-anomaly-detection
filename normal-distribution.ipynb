# Illustration for Normal Distribution

```python
import json

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('config.json', 'r') as config_file:
    config = json.load(config_file)
```

```python
BASE_COLOR = config['base_color']
LIGHT_COLOR = config['light_color']
DPI = config['dpi']
EDGE_COLOR = config['edge_color']
AXIS_COLOR = config['axis_color']
GRID_COLOR = config['grid_color']
FORMAT = config['format']
```

```python
palette = f"blend:{BASE_COLOR},{LIGHT_COLOR}"
```

```python
rc = {'axes.edgecolor': AXIS_COLOR,
      'axes.labelcolor': AXIS_COLOR,
      'grid.linestyle': '--',
      'text.color': AXIS_COLOR,
      'xtick.color': AXIS_COLOR,
      'ytick.color': AXIS_COLOR,
      'grid.color': GRID_COLOR}
```

```python
sns.set(font_scale=2)
sns.set_style("whitegrid", rc=rc)
sns.set_context("talk")
```

```python
np.random.seed(42)
```

```python
LINEWIDTH = 2
```

```python
df = pd.DataFrame(np.random.normal(0, 1, 100000), columns=['данные'])
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(df, x='данные', bins=100, color=BASE_COLOR, edgecolor=EDGE_COLOR)
ax.set(ylabel='')
plt.axvline(3, linestyle='--', color=AXIS_COLOR, linewidth=LINEWIDTH)
plt.axvline(0, linestyle='--', color=AXIS_COLOR, linewidth=LINEWIDTH)
plt.axvline(-3, linestyle='--', color=AXIS_COLOR, linewidth=LINEWIDTH)
plt.savefig(f"images/normal_distribution.{FORMAT}", bbox_inches='tight', transparent=True, dpi=DPI)
```
