# Illustration for Normal Distribution

```python
import json

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('config.json', 'r') as config_file:
    config = json.load(config_file)
```

```python
BASE_COLOR = config['base_color']
LIGHT_COLOR = config['light_color']
DPI = config['dpi']
EDGE_COLOR = config['edge_color']
```

```python
palette = f"blend:{BASE_COLOR},{LIGHT_COLOR}"
```

```python
np.random.seed(42)
```

```python
LINEWIDTH = 2
```

```python
sns.set(font_scale=2)
sns.set_style("whitegrid")
sns.set_context("talk")
```

```python
df = pd.DataFrame(np.random.normal(0, 1, 100000), columns=['данные'])
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(df, x='данные', bins=100, color=BASE_COLOR, edgecolor=EDGE_COLOR)
ax.set(ylabel='')
plt.axvline(3, linestyle='--', color='black', linewidth=LINEWIDTH)
plt.axvline(0, linestyle='--', color='black', linewidth=LINEWIDTH)
plt.axvline(-3, linestyle='--', color='black', linewidth=LINEWIDTH)
plt.savefig("images/normal_distribution.png", bbox_inches='tight', transparent=True, dpi=DPI)
```
